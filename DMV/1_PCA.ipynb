Problem Statement:
Feature Transformation -
A. To use PCA Algorithm for dimensionality reduction.
You have a dataset that includes measurements for different variables on wine
(alcohol, ash, magnesium, and so on). Apply PCA algorithm & transform this data
so that most variations in the measurements of the variables are captured by a small
number of principal components so that it is easier to distinguish between red and
white wine by inspecting these principal components.

```python
#Libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#Dataset
data_url = "https://media.geeksforgeeks.org/wp-content/uploads/Wine.csv"
data = pd.read_csv(data_url)
```

```python
X = data.drop('Customer_Segment', axis=1)
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
num_components = 2 
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Plot transformed data based on the first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['Customer_Segment'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Transformed Data: Wine Segments')
plt.show()
```

```python
num_components = 3
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(X_scaled)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=data['Customer_Segment'], cmap='viridis')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
```
